##Spatial Frequency

Spatial frequency can be expressed in terms of 'events' per angular unit (events/radian) or distance units (events/m). Angular spatial frequency is converted to distance spatial frequency by the definition of angle $\theta = S/R$ where $S$ is the arc length along a circle and $R$ is the radius of the circle. To convert from distance spatial frequency to angular spatial frequency divide by $R$, the distance over which the conversion is performed.  This $R$ can be the focal length of an optical system if the distance is on the optical focal/image plane. $R$ can also be the distance from the sensor to the object plane if the distance is from the sensor to the object plane.  

The 'events' can be any observable thing, e.g. bricks, cars, windows, etc. When doing engineering it is convenient to think of an event as a cycle, where the cycle can again have almost any meaning. The most common definition of cycle  in this context is to consider two sub-events, such as a positive and negative half cycle of a sinusoid.  The full cycle is then the full 2$\pi$ extent of a sinusoid.  It is however important to note that there is no special meaning in the use of cycle here, it is merely a convenient 'event' to be used.

For the initial analysis below, the spatial frequency will be in angular units, cycles/radian. Towards the end of the analysis the units will be converted to pixel counts.

##Aerosol Models

The atmospheric aerosol particle size distribution covers a relatively wide range. Assuming  homogeneous spherical particles, the particle size distribution (PSD) approximated by the lognormal probability law,
$$
f(a)=\frac{1}{\sigma (a/2) \sqrt{2\pi}}\exp\left\{-\frac{\ln^2(a/a_0)}{2\sigma^2}\right\}
$$
where $a$ is the particle diameter and $\sigma^2$ is the lognormal distribution variance. 

For clear sky atmospheric molecules $a\lt 0.01$ $\mu$m, while for good visibility aerosol $0.1 \lt a\lt0.5$ $\mu$m. For larger (smoke and dust) aerosol the particle size can increase from 2 to 4 $\mu$m (Ref 1, Ref 3, Ref 5). In dusty aerosol the particle sizes can be even bigger, but such aerosol settles out of the atmopsphere relatively quickly. 

The particle size distribution usually is relatively wide. Quite often the particle distributions are multi-mode as well, meaning that more than one particle distribution is present in the atmosphere at the same time.

## Optical Thickness, Transmittance and Angstrom Exponent

The transmittance of an optical medium can be characterised by the medium's extinction coefficient $\gamma$. 
If the medium along the path is inhomogeneous, and the variation of $\gamma$ along the path is known, the attenuation coefficient  can be written as a normalizing constant $\gamma_o$ multiplied by a path-dependent function $f(x)$, $\gamma(x)=\gamma_o f(x)$. The function $f(x)$ represents the profile of $\gamma$ along the optical path. 
The transmittance of the path in an inhomogeneous medium is then
$$
\tau(R)
 =\exp\left[-\int_{0}^{R}\gamma_o f(x) dx\right]\nonumber
 =e^{-\gamma_o\overline{R}}
$$
where
$$\overline{R} = \int_0^R{f(x) dx}$$

The optical thickness  of the medium is defined as the exponent in the previous equation, $\beta=\gamma R$. If the optical thickness is small, the transmittance is high and the medium is clear. On the other hand, if $\beta$ is high, transmittance is low and the medium is opaque. 


The attenuation through the atmosphere comprises scattering and absorption.  The attenuation varies with wavelength and location in the atmosphere. Assuming a stratified atmosphere where the attenuation coefficient varies only with height $h$ above ground and not on lateral displacement, the attenuation coefficient can be written in terms of the spectral scattering coefficient $\sigma_\lambda(h)$ and spectral absorption coefficient $\alpha_\lambda(h)$, $\gamma_\lambda(h) =\sigma_\lambda(h) + \alpha_\lambda(h)$, where $h$ is height above ground.

Most of the atmospheric aerosol is contained in the boundary layer, assuming a downward view path and uniform aerosol distribution in this boundary layer it follows that $\beta=\sec(\phi)\gamma H$ where $\phi$ is the view zenith angle and $H$ is the thickness of the boundary layer (or the observer altitude if inside the boundary layer).  

The optical thickness for continental aerosol  varies with wavelength, often modelled with the Angstrom extinction law
$$
\beta_\lambda = \beta_{1 \mu{\rm m}}\lambda^{-\alpha}$$
where $\alpha$ isthe Angstrom exponent.
A typical range for $\alpha$ is 0.5-2.5, with an average for natural atmospheres of around 1.3±0.5 (Ref 6).  The Angstrom exponent is inversely related to the average size of the particles in the aerosol: the smaller the particles, the larger the exponent. Thus, Angstrom exponent is a useful quantity to assess the particle size of atmospheric aerosols or clouds, and the wavelength dependence of the aerosol/cloud optical properties. 
Larger values of $\alpha$ imply a relatively high ratio of small particles to large  particles.
Cloud droplet, usually with large sizes and thus very smaller Angstrom exponent (nearly zero), is spectrally neutral, which means, e.g., the optical depth does not change with wavelength.

##Scattering by a Particle

The angular distribution of light scattering around a particle is described by the phase function $p(\theta)$. The value of $p(\theta)d\Omega/4\pi$ gives the  probability of light scattering in the solid angle $d\Omega = \sin\theta d\theta d\phi$  where $\theta$ and $\phi$ are spherical angles. It is difficult to calculate the phase function and a simplified (but not very accurate) approximation, called the Heney-Greenstein phase function, is sometimes used.  It has been shown sufficiently that the Heney-Greenstein phase function is a poor approximation for the complexities of real-world aerosol, comprising multiple particle mode distributions (e.g. molecules, fine particles and coarse particles). Each of these particle modes has a different phase function and the effective phase function is therefore a combination of individual phase functions. ##Diffraction Around a Particle
By Babinet's principle the diffration through an aperture and around a solid object of the same shape is the same (Ref 2 p196; Ref 4 p 381). This diffraction total angular spread is approximately given by the twice the radius to the first zero in the diffraction pattern of a circular aperture $\theta\approx 1.22\lambda/a$ radians (Airy disk, Ref 4 p397), where $\lambda$ is the wavelength and $a$ is the diameter of the aperture (or the size of the particle in this analysis).

Visualise a (laser) beam of light being diffracted around the particle. This beam is spread out into a cone of angle $1.22\lambda/a$ radians. 
Consider the half the diffraction angle to be half a cycle, then the spatial frequency corresponding to the diffraction pattern is given by $f_c = 0.5 a/(0.61\lambda)\approx a/\lambda$ cycles/radian.

##Modulation Transfer Function for Aerosol Scattering

In the visible spectrum there is no strong molecular absorption and the absorption present is mainly aerosol absorption. For this current analysis, the aerosol absorption is assumed to be small - this is not a valid assumption for larger aerosol particles, particularly wet aerosol; and this assumption is also not valid in the infrared spectrum.

Ref 3, Chapter 4 provides a detailed (yet quite incomprehensible) analysis of the derivation of the atmospheric point spread funtion. Ref derives the MTF for small angles (paraxial) around the incident beam, applicable to predominantly forward scatter as for large/coarse aerosols.  This derivation therefore does not apply to optically thick media (Ref 3, p90). Equation 4.55 in Ref 3 shows that the MTF has a gaussian shape for small frequencies
$$
S(\omega) = \exp\left\{
-
\beta_\lambda+\beta_\lambda\omega_0\left(
1-\frac{\omega^2}{12a^2}\right)\right\}$$
where $\omega$ is a normalised frequency. If $1<\frac{\omega^2}{12a^2}$ at high frequencies, the MTF becomes
$$
S(\omega) = \exp\left\{
-
\beta_\lambda-
\beta_\lambda\omega^2\left(\frac{\omega_0}{12a^2}\right)\right\}$$
The variable $a$ stems from the phase function used in the derivation $p(\theta) = 4a^2\exp(-a^2\theta^2$).
The meaning of the $\omega_0$ variable is defined (Ref 3, p 17) as the single scatter albedo defined as the ratio of scattering coefficient to the total extinction coefficient $\omega_0 =\sigma_\lambda/(\sigma_\lambda-\alpha_\lambda)$. Values for $\omega_0$ varies from 0.78-0.88 for burning boimass, 0.83-0.9 for Mexico City, and 1 for scattering with no absorption.  Hence, 

$$
S(\omega) = \exp\left\{
-
\beta_\lambda-
\sigma_\lambda\omega^2\left(\frac{1}{12a^2}\right)\right\}$$



Kopeika's model equation has a number of variations, but with more or less the following general form (Equation 9, in Ref 1)

\begin{eqnarray}
M_a(f_r)
&=&\exp\left(
-\beta\left(\frac{f_r}{f_c}\right)^2
\right) \;\;{\rm for}\; f_r \lt f_c\\\
&=&\exp\left(
-\beta
\right) \;\;{\rm for}\; f_r \ge f_c\
\end{eqnarray}

where $\beta = \sec\phi\int_{h_1}^H \gamma_\lambda(h) dh$, $f_r$ is spatial frequency,  $h_1$ is the target-side end of the optical path, $H$ is the sensor-side end of the path, $\theta$ is slant angle; accounting for longer path length at slant angles, i.e. the path length is $R=\sec\theta\int_{h_1}^H dh$. The normalising frequency $f_c$ scales the width of the frequency-dependent component.  The form of the first part of Kopeika's equation is similar to the equation derived in Ref 3. With $\beta = \sec\phi\int_{h_1}^H \gamma_\lambda(h) dh$. Under the assumption that absorption is small, 



\begin{eqnarray}
M_a(f_r)
&=&\exp\left[
-\sigma_\lambda\left(\frac{f_r}{f_c}\right)^2
\right] 
\end{eqnarray}

The normalising mapping between Ref 3 and Ref 1 frequency is not evident. It is not clear how to move from Ref 3 frequency to the Ref 1 definition, which is normalised with respect to the particle size (i.e. the diffractive effect of the particle).

##Numerical Analysis

Using the Kopeika formulation, the MTF is next calculated for a  number of optical depths and particle sizes.

```python
def MTFlin(od, ps, wavelength, spatFreq):
    fcc = ps/wavelength
    MTFFS = np.exp(-od *(spatFreq/fcc) ** 2)
    MTFMS = np.exp(-od)
    MTF = np.maximum(MTFFS,MTFMS)
    return MTF
    
optDepth = [ .1, 1, 3]
partSize = np.asarray([.5, 1.0, 4]) * 1e-6
wavelength = 0.5e-6
fc = partSize[0]/wavelength
# spatFreq = np.linspace(np.log10(0.01), np.log10(10 * fc), 1000)
spatFreq = np.linspace(0, np.sqrt(2) * 20 * fc, 500)

p = ryplot.Plotter(2,figsize=(12,7))
for od in optDepth:
    for ps in partSize:
        MTF = MTFlin(od, ps, wavelength, spatFreq)
        label = 'opt.depth={} part.size={} $\mu$m'.format(od,ps*1e6)
        p.semilogX(1,spatFreq,MTF,xlabel='Spatial frequency [cy/rad]',
                   ylabel='MTF',label=[label])
```

A fundamental problem working with transforms is the 'dynamic resolution' when converting from the one  domian to the other. If you require good resolution in both frequency and spatial domains, you need very large data sets, in order to obtain sufficient resolution in both domains.  This problem is especially significant when working with two-dimensional data. In this particular case, the challenge is sufficient resolution to accurately model the spike at the centre of the psf, while also modelling the psf at longer distances from the centre, while keeping the data set size within hardware constraints.

The numerical analysis immediately following demonstrates this problem: the frequency range must be very high in order to produce a sharp spike in the spatial domain. After some stumbling against the limitations of the numerical process, it suddenly became evident that this specific problem can be solved analytically:  All the terms can be formulated in simple mathematics. But first the numerical attempt. 

```python
def CalculateMtfPsf(od, ps, wavelength, fc,fcSize):
    # Fill a two-dimensional array with the rotationally symmetrical MTF
    fcmax = fc * 100
    xf,yf = np.mgrid[-fcmax:fcmax:np.complex(0,fcSize), -fcmax:fcmax:np.complex(0,fcSize)]
    
    #map frequency increments to spatial increments after ft
    freqInterval = xf[1][0] - xf[0][0]
    spatInterval = 1 / (fcSize * freqInterval)
    spmax = 1 / freqInterval
    xd,yd = np.mgrid[-spmax:spmax:np.complex(0,fcSize), -spmax:spmax:np.complex(0,fcSize)]

    print('Full data set has {0} samples:'.format( fcSize))
    print('  Full frequency range {0} with intervals {1} cy/rad'.format(fcSize*freqInterval, freqInterval ))
    print('  Full angular range {0} with intervals {1} rad (or rad/cy)'.format(fcSize*spatInterval, spatInterval ))
    
    fcr = np.sqrt(xf**2 + yf**2)
    MTFr = MTFlin(od, ps, wavelength, fcr)
    
    #shift origin to corners in order to do fft, do fft and then shift back to centre
    psfr = numpy.fft.fftshift(numpy.fft.ifft2(numpy.fft.ifftshift(MTFr)))
      
    return xf,yf, MTFr, xd,yd, psfr
```

```python
def PlotMtfPsf(fcSize,unit, xf,yf, MTFr, xd,yd, psfr,csize):
    # select the central portion for plotting
    fcentre = (fcSize-1)/2

    fl = fcentre-csize
    fu = fcentre+csize

    q = ryplot.Plotter(3,1,2,figsize=(20,8))
    q.meshContour(1,xf[fl:fu,fl:fu],yf[fl:fu,fl:fu],MTFr[fl:fu,fl:fu],
                  xlabel='Spatial frequency [cy/{}]'.format(unit), 
                  ylabel='Spatial frequency [cy/{}]'.format(unit),
                  cbarshow=True)
    q.meshContour(2,xd[fl:fu,fl:fu],yd[fl:fu,fl:fu],psfr[fl:fu,fl:fu],
                  xlabel='Angle [{}]'.format(unit), 
                  ylabel='Angle [{}]'.format(unit), cbarshow=True)
    
    qq = ryplot.Plotter(4,1,2,figsize=(20,8))
    qq.plot(1,xf[fl:fu,fcentre],MTFr[fl:fu,fcentre],
                  xlabel='Spatial frequency [cy/{}]'.format(unit), 
                  ylabel='MTF')
    qq.plot(2,xd[fl:fu,fcentre],psfr[fl:fu,fcentre],
                  xlabel='Angle [{}]'.format(unit), ylabel='Angle [rad]')
```

```python
fcSize = 1001
xf,yf, MTFr, xd,yd, psfr = CalculateMtfPsf(od, ps, wavelength, fc,fcSize)
PlotMtfPsf( fcSize,'rad', xf,yf, MTFr, xd,yd, psfr, csize = 100)
```

##Analytical Analaysis

The analysis was done separately. Kopeika's equation can be reformulated as
\begin{equation}
M(\rho)=
\exp\left(-\beta\right) + \left[\exp\left\{-\beta\left(\frac{\rho}{f_c}\right)^2\right\}- \exp\left(-\beta\right) \right]
{\rm cyl}\left(\frac{\rho}{f_c}\right)
\end{equation}
where 
The cyl  function is  defined as follows:
\begin{eqnarray}
{\rm cyl}\left(\frac{r}{d}\right)
&=& 1 \;\;{\rm for}\; 0\le r < d/2\nonumber\\
&=& 0.5 \;\;{\rm for}\; r= d/2\nonumber\\
&=& 0 \;\;{\rm for}\; r> d/2\nonumber
\end{eqnarray}


Applying the Hankel transform (documented separately), the following result is obtained for the point spread function in the spatial domain
\begin{eqnarray}
m(r)&=&
\exp\left(-\beta\right)\frac{\delta(r)}{r\pi} - 
\exp\left(-\beta\right)|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) 
+\frac{\pi}{\beta} |f_c|^2\,\exp\left\{-\pi^2 (f_c r)^2/\beta\right\}\star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) 
\end{eqnarray}

where the Sombrero function is defined as
\begin{equation}
{\rm somb}\left(\frac{r}{d}\right)=
\frac{2J_1\left(\frac{\pi r}{d}\right)}{\left(\frac{\pi r}{d}\right)}
\end{equation}
and 
$J_1$ is the first-order Bessel function of the first kind.


The SciPy documentation for Bessel functions are confusing: but note that scipy.special.jn is simply an alias for scipy.special.jv. SciPy Bessel functions take two arguments. The first is the order and the second is the parameter. For example, $J_v(z)$ is evaluated using scipy.special.jn(v, z).


##Numerical Implementation of the Analytic Solution
Implementing the $\rm cyl$, $\rm somb and $\rm gaussian$ functions in numerial terms is relatively easy. The challenge arises when implementing the Dirac-delta impulse.  The Dirac-delta has the unique property that it has zero width and infinite height (both of which are impossible to implement numerically) but unity area in 1-D and volume in 2-D.  In the code below, the Dirac-delta is implemented as the limiting case of a gaussian shape, with very, very narrow width, thereby approaching the Dirac-delta impulse.

```python
def somb(x,y,d):
    """Calculate the 2-D sombrero function J1(pi.fc/d)/(pi.fc/d) where 
    J1 is the first order Bessel function of the first kind.
    x, y, are the sample points and d is the radius of the first zero
    """
    arg = np.pi * np.sqrt(x**2 + y**2) / d
    #the strange '+ (arg==0)' below is to handle divide by zero, if so set to 1
    s = (2 * scipy.special.jv(1,arg)/(arg + (arg==0))) * (arg!=0) + 1 * (arg==0)
    return s
         
```

```python
def gaussian(x,y,d,a):
    """Calculate the 2-D gaussian function exp(-(pi/a) (r/d)^2) where 
    x, y, are the sample points and (d, a) determine width of the shape
    """
    arg = (np.pi/a) * (np.sqrt(x**2 + y**2) / d ) ** 2
    b = np.exp(-arg)
    return b
```

```python

def AnalyticPSF(od, ps, fc, wavelength, fcSize, spmax,unit,impulseFc, rscale, pltaxis):

    strCase = 'Part.size={} m, opt.depth={} fc={} cy/rad rscle={}'.format(ps,od,fc,rscale)
    spmax *= rscale
    fc /= rscale
    impulseFc /= rscale
    
    xs,ys = np.mgrid[-spmax:spmax:np.complex(0,fcSize), -spmax:spmax:np.complex(0,fcSize)]
    s = fc * fc * np.pi * somb(xs, ys, 1/fc) / 4
    g = fc * fc * np.pi * gaussian(xs, ys, 1/fc, od/np.pi) / od
    s2 = -np.exp(-od) * s
    #fftconvolve is faster than convolve2d
    #why do we have to scale with the linear size? seems ok, but why?
    sg = scipy.signal.fftconvolve(g,s,mode='same')/ (fcSize /(rscale ** 2) ) 
    
    #impulseFc  how large should this be?
    imp = impulseFc * impulseFc * gaussian(xs, ys, 1/impulseFc, 1) 
    
    sum = sg + s2 + imp
    
    fcentre = (s.shape[0]-1)/2
    
    q = ryplot.Plotter(5,3,2,figsize=(14,20), figuretitle=strCase)
    q.meshContour(1,xs,ys,s,ptitle='somb',
                  xlabel='Spatial [{}]'.format(unit), 
                  ylabel='Spatial [{}]'.format(unit), cbarshow=True,contourLine=False)
    q.meshContour(2,xs,ys,g,ptitle='gauss',
                  xlabel='Spatial [{}]'.format(unit), 
                  ylabel='Spatial [{}]'.format(unit), cbarshow=True,contourLine=False)
    q.meshContour(3,xs,ys,sg,ptitle='somb**gauss',
                  xlabel='Spatial [{}]'.format(unit), 
                  ylabel='Spatial [{}]'.format(unit), cbarshow=True,contourLine=False)
    q.meshContour(4,xs,ys,s2,ptitle='somb2',
                  xlabel='Spatial [{}]'.format(unit), 
                  ylabel='Spatial [{}]'.format(unit), cbarshow=True,contourLine=False)
    q.meshContour(5,xs,ys,imp,ptitle='imp',
                  xlabel='Spatial [{}]'.format(unit),
                  ylabel='Spatial [{}]'.format(unit), cbarshow=True,contourLine=False)
    q.meshContour(6,xs,ys,sum,ptitle='somb2+somb**gauss',
                  xlabel='Spatial [{}]'.format(unit), 
                  ylabel='Spatial [[{}]'.format(unit), cbarshow=True,contourLine=False)
     
    qq = ryplot.Plotter(6,1,1,figsize=(14,10), figuretitle=strCase)
    qq.plot(1,xs[:,fcentre],s[:,fcentre],label=['somb'])
    qq.plot(1,xs[:,fcentre],g[:,fcentre],label=['gauss'])
    qq.plot(1,xs[:,fcentre],sg[:,fcentre],label=['convolution'])
    qq.plot(1,xs[:,fcentre],s2[:,fcentre],label=['somb2'])
    qq.plot(1,xs[:,fcentre],imp[:,fcentre],label=['imp'])
    qq.plot(1,xs[:,fcentre],sum[:,fcentre],label=['sum'],pltaxis=pltaxis,
           xlabel='Spatial [{}]'.format(unit), ylabel='Magnitude',plotCol=['k'])

```

##Satellite Sensor
The sensor pixel instantaneous field of view is determined by the pixel size and the focal length. To convert spatial frequency $f_r$ in cy/rad to $f_m$ cy/m divide by the focal length $f_m = f_r/f$ with units as shown below. To now further convert from cy/m to cy/pixel, divide by pixels/m to get cy/pixel. This last step is the same as to multiply by m/pixels (which is also the pixel size ($d$)).
$$
\left[\frac{{\rm cycles}\;{\rm m}_{\rm f}}{{\rm m}_{\rm d}}\right]
\left[\frac{1}{ {\rm m}_{\rm f}}\right]
\left[\frac{{\rm m}_{\rm d}}{{\rm pix}}\right]
= \frac{{\rm cycles}}{{\rm pixel}}
 $$
Hence to convert spatial frequency from cycles/rad to cycles/pixel use $f_p = d f_r/f = f_r \cdot{\rm IFOV}$. 
Conversely to convert angle/distance from rad/cycle to pixels/cycle use the inverse.

Now calculate the point spread function in the focal plane of a sensor. In this case the frequency axis scales with focal length and detector pixel size, to arrive at pixel displacement (instead of angles as before).

##Truncating the Point Spread Function
The graph in the previous section shows that the point spread function will have to be well larger than (50 million)^2 pixels!  This is clearly not practical.  Any real implementation will have to truncate the PSF convolution kernel to much smaller. The primary criterium for determining the kernel size is to decide how wide the aerosol PSF adjacency effect must be in the specifica application. Suppose the requirement is that the adjacency of targets should be effective at 1 degrees _outside_ the field of view, then the PSF convolution kernel must be at least 2 degrees wide, from edge to edge.  In the sensor example above this would require a kernel of 58000 pixels on each side, which is a huge computing challenge - to the point of being impractical (just the kernel alone would require 28GB of memory).  Alternative strategies could be to break the single large kernel down in to a series of smaller kernels, to be applied many times.

Truncating the  PSF would mean that the central portion of the 2-D suface calculated above would be used, suitably subsampled at the sensor pixel intervals. In all practical cases, this subsampled region would have an almost constant value (except for the impulse at the central position).  

If the PSF is truncated, a windowing function should be used to minimise the truncation error effects.  See https://en.wikipedia.org/wiki/Window_function and http://www.sandv.com/downloads/0301wick.pdf.

##Practical Approach

Reflection on the resuls derived above very quickly leads to the conclusion that is it impractical to use the convolution (or frequency domain multiplication) approach to model aerosol point spread function for sensors with small fields of view.  The sensor resolution and kernel and input image sizes forces a solution that is not readily solved by numerical means. The following observations are made:

1.  If the field of view is comparable to the aerosol diffraction or scattering angle (within an order of magnitude), as shown in the small camera example above, the point spread function is not impossibly large (yet still quite large).  In this case, the magnitude of the of the point spread function varies significantly over a thousand or two pixels. The magnitude shape of the point spread function is therefore important over the size of the point spread function and image.  In this case, a point spread function with the correct shape must be calculated and convolved with the image to account for adjacency effects.

2.  If the field of view is considerably smaller than the aerosol diffraction or scattering angle (several orders of magnitude), as shown for the satellite camera above, the point spread function becomes too wide to compute and convolve numerically. Consider the example above where the diffraction/scattering covers one or two radian in angular width. In this case the adjacency effect remains significant for several tens of thousands of pixels away.  A different solution must then be found. 

    1.    If the scene statistical content at sensor FOV scope is stationary over the larger scene, the scattering/adjacency effect of any part in the scene will contribute to any other part in the scene is equal amounts (large scale stationarity).
    
    2.    If the field of view is small compared to the width of the central lobe in the point spread function, the scattering point spread function can be considered approximately constant (in the central and truncated part), except for the impulse at the origin. Convolution with a kernel with contstant value kernel (except for the impulse) is simply the same as adding up all the values and multiplying with the constant value.
    
    3.    Considering the above two points, the aerosol adjacency effect for an extremely narrow field of view sensor can be modelled by simply adding all the pixels in the image prior to adjacency modelling and multiplying the sum with the value of the kernel near the centre (excluding the impulse) and then add this to the original image (representing the impulse in the kernel) but slightly scaled down to account for the kernel integral required to be unity for a no-gain kernel.


##Analytical Derivation


The cyl and somb functions are defined as follows (Ref 7):

$$\begin{eqnarray}
{\rm cyl}\left(\frac{r}{d}\right)
&=& 1 \;\;{\rm for}\; 0\le r < d/2\nonumber\\
&=& 0.5 \;\;{\rm for}\; r= d/2\nonumber\\
&=& 0 \;\;{\rm for}\; r> d/2\nonumber
\end{eqnarray}
$$

$$\begin{equation}
{\rm somb}\left(\frac{r}{d}\right)=
\frac{2J_1\left(\frac{\pi r}{d}\right)}{\left(\frac{\pi r}{d}\right)}
\end{equation}
$$

where $J_1$ is the first-order Bessel function of the first kind.

The Dirac-delta function $\delta(r)$ has infinite height, zero width and the area/volume under the impulse response is one,

$$\begin{equation}
\int_0^\infty\frac{\delta(r)}{r\pi}dr = 1.
\end{equation}
$$

This is important when the psf is implemented in discrete, pixelated form.

###Derivation
The original Kopeika aerosol MTF formulation is as follows:
$$
\begin{eqnarray}
M(\rho)
&=&\exp\left(-\beta\left(\frac{\rho}{f_c}\right)^2\right) 
   \;\;{\rm for}\; \rho < f_c\nonumber\\
&=&\exp\left(-\beta\right) 
   \;\;{\rm for}\; \rho \ge f_c\
\end{eqnarray}
$$

This can be reformulated as
$$
\begin{equation}
M(\rho)=
\exp\left(-\beta\right) + \left[\exp\left\{-\beta\left(\frac{\rho}{f_c}\right)^2\right\}- \exp\left(-\beta\right) \right]
{\rm cyl}\left(\frac{\rho}{f_c}\right).
\end{equation}
$$

Applying the Hankel transform

$$\begin{eqnarray}
m(r)&=&
\exp\left(-\beta\right)\frac{\delta(r)}{r\pi}  
+\left[\frac{\pi}{\beta} |f_c|^2\,\exp\left\{-\pi^2 (f_c r)^2/\beta\right\}- \exp\left(-\beta\right)\frac{\delta(r)}{r\pi} \right]
\star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) \nonumber\\
&=&
\exp\left(-\beta\right)\frac{\delta(r)}{r\pi}  
+\frac{\pi}{\beta} |f_c|^2\,\exp\left\{-\pi^2 (f_c r)^2/\beta\right\}\star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) 
- \exp\left(-\beta\right)\frac{\delta(r)}{r\pi} \star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) \nonumber\\
&=&
\exp\left(-\beta\right)\frac{\delta(r)}{r\pi}  
- \exp\left(-\beta\right)\frac{\delta(r)}{r\pi} \star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) 
+\frac{\pi}{\beta} |f_c|^2\,\exp\left\{-\pi^2 (f_c r)^2/\beta\right\}\star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r). \nonumber
\end{eqnarray}
$$

Which leads to the final formulation

$$
\begin{eqnarray}
m(r)&=&
\exp\left(-\beta\right)
\left[
\frac{\delta(r)}{r\pi} - 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) \right]
+\frac{\pi}{\beta} |f_c|^2\,\exp\left\{-\pi^2 (f_c r)^2/\beta\right\}\star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r) 
\end{eqnarray}
$$

because

$$
\begin{equation}
\frac{\delta(r)}{r\pi} \star\star\, 
|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r)
=|f_c|^2\left(\frac{\pi}{4}\right){\rm somb}(f_c r)
\end{equation}
$$

where

$$
\begin{equation}
{\rm somb}\left(f_c r\right)=
\frac{2J_1\left(\pi f_c r\right)}{\left(\pi f_c r\right)}
\end{equation}
$$
